```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("nordic-thingy52.csv")
```

```python
for distance in df.Actual_Distance.unique():
    df_distance = df[df.Actual_Distance == distance]
    print("Distance: {}".format(distance))
    print(df_distance.describe())
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style("whitegrid")
sns.boxplot(x = df["Actual_Distance"], y = df["Calculated_Distance"])
plt.title("Calculated Distance Quartiles by Actual Distance\n", fontsize=22)
plt.xlabel("Actual Distance in Meters")
plt.ylabel("Calculated Distance in Meters")
plt.show()
```
